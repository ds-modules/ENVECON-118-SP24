#### Summarizing Variables

You can view basic summary statistics of a variable by using `summary()` or `summarise()` functions. The first function, from base package, is more generic, while the latter function, from `dplyr` package, allows us to customize which summary statistics we would like to calculate. 

<u>TRY</u> using the function `summary()` on the variable "logPrice".

```r
summary(sec1_df$logPrice)
```

<u>TRY</u> using the function `summarise()` to calculate min, max, and range of the variable "logPrice". (Note: `summarise()` function comes from package 'dplyr' which must be loaded prior to use.)

```r
summarise(sec1_df, "Min" = min(logPrice), "Max" = max(logPrice), "Range" = max(logPrice) - min(logPrice))
```
